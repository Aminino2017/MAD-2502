```python
#class for a polynomial
class Polynomial:
    '''univeraite polynomials'''
    
    def _remove0s(self):   #convention is that _functionname means that functionname is not meant to be called by users
        '''deletes leading 0-coefficients'''
        
        try:
            while self.coeffs[-1]==0:
                del self.coeffs[-1]
        except IndexError:   #returns empty list
            pass
        
        
    def __init__(self,coeffs,var='x'):
        '''initialize a polynomial, constants furtherst left, varaibles furthest right'''
        from numbers import Number
        '''NUmber is overarching class of all numbers, complex, real, floats, integers'''
        
        if any(not isinstance(i,Number) for i in coeffs):
            raise TypeError('only numbers allowed')
            
        self.coeffs=coeffs[:]
        self._remove0s()
        self.var=var[:]
        self.deg=len(self.coeffs)-1   #degree of 0-polynpomials is -1
        
    def __eq__(self,other):
        '''equality of polynomials'''
        
        return (self.var==other.var and self.coeffs==other.coeffs)

    
    def __add__(self,other):
        '''add two polynomials'''
        
        if self.var!=other.var:
            raise ValueError('variables must be indentical')
            
        my_c,other_c=self.coeffs[:],other.coeffs[:]           
        
        while len(my_c)<len(other_c):
            my_c.append(0)
        while len(my_c)>len(other_c):
            other_c.append(0)
        
        sump= [sum(x) for x in zip(my_c,other_c)] 
        
        return Polynomial(sump,self.var)
    
    def __sub__(self,other):
        if self.var!=other.var:
            raise ValueError('variables must be identical')
        
        negated_coeffs=[-i for i in other.coeffs]
        return self+Polynomial(negated_coeffs, self.var)

    def __mul__(self,other):
        if self.var!=other.var:
            raise ValueError('variables must be identical')
        coeffs=[]
        result_degree=self.deg+other.deg
        a=self.coeffs +(result_degree-self.deg)*[0]
        b=other.coeffs +(result_degree-other.deg)*[0]
        
        for k in range(0,result_degree +1):
            coeffs.append(sum([a[i]*b[k-i] for i in range(k+1)]))        
        
        return Polynomial(coeffs,self.var)
    
    
    def _derivative(self):
        deriv=self.coeffs   #hold onto the coefficients
        for i in range(len(deriv)+1):    #multiply each coefficient by the power of the variable
            deriv[i]*=i
        del deriv[0]    #remove initial constant
        return deriv
        
    
    
    def __str__(self):
        strings=[]    #will have final result
        var=self.var  #holds varaible of polynomial
        check=self.coeffs   #looks for 0 coefficients
        signs=[]       #holds sign of coefficient
        deg=[]       #hold degree of particular polynomial
        

                
        for i in range(len(check)):
            if check[i]> 0 and not check[0]:
                signs.append('+')
            if check[i]<0:
                signs.append('-')
        for i in range(0, len(check)):
            deg.append(i)
            
            
        for i in range(len(check)):
            if check[i]==0:   #if aprticular value has 0 coefficient
                del check[i] 
                
                
        #for i in range (len(check)):
         #   for j in range (2, len(check))
          #      strings.append(check[i],var,'^',deg[j])
        
        
    
    
    def evaluate(self,x):
        '''evaluate the polynomial at x'''  #this is known as Horner's scheme
        a,d=self.coeffs, self.deg
        if a==[]:
            return 0
        final=a[d]
        for i in range(d,0, -1):
             final=a[i-1]+x*final
        return final
```

```python
def interpolate(x,y):
    '''find the polynomial interpolating (x[i],y[i])'''
    k=len(x)-1
    ls=[]
    
    for j in range(k+1):
        # compute \ell_j(x)
        ell=Polynomial([1])
        for m in range(k+1):
            if m!=j:
                denom=x[j]-x[m]
                ell*=Polynomial([-x[m]/denom, 1/denom])
                
        ls.append(ell)
    L=Polynomial([0])
    for j in range(k+1):
        L+=ls[j]*Polynomial([y[j]])
    return L
```

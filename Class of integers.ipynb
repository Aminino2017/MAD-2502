Write a Python class TinyNumber as follows:

Instances of this class are generated by providing an integer x such that 0<=x<=999 e.g., we can write a=TinyNumber(42

Provide an implementation of __str__ , returning a representation of the number as English text. E.g., print(TinyNumber(12)) should return the string "twelve".

Provide an implementation of +, adding two TinyNumbers modulo 1000.

Provide an implementation of == to check if two TinyNumbers are equal.

```python
class TinyNumber:
    
    NumD0={0:'zero'}
    NumD1s={1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight', 9:'nine'}
    NumD10_19s={10:'ten', 11:'eleven', 12:'twelve', 13:'thirteen', 14:'fourteen', 15:'fifteen', 16:'sixteen', 17:'seventeen', 18:'eighteen', 19:'nineteen'}
    NumD10s={20:'twenty', 30:'thirty', 40:'forty', 50:'fifty', 60:'sixty', 70:'seventy', 80:'eighty', 90:'ninety'}
    NumD100s={100:"one hundred", 200:"two hundred", 300:"three hundred", 400:"four hundred", 500:"five hundred", 600:"six hundred", 700:"seven hundred", 800:"eight hundred", 900:"nine hundred"}

    def __init__(self,value):
        '''Will define each value added to the class'''
        self.value=value
        self.l=len(str(value))
        
        
    def __str__(self):
        '''convert digits into text'''
        a=self.value
        t=self.l
        if a==0:
            return NumD0[0]
        if 0<a<10:
            return NumD1s[a]
        if 9<a<20:
            return NumD10_19s[a]
        if 19<a<1000:
            if a in NumD10s:
                return NumD10s[a]
            if a in NumD100s:
                return NumD100s[a]
            else:
                if t==2:
                    r=a%10  #will hold second digit
                    n=a-r   #removing second digit gives 10s place
                    first=NumD10s[n]
                    second=NumD1s[r]
                    return first+'_'+second
                
                else:
                    r=a%100  #will hold tens place
                    n=r%10   #will hold ones place
                    s=a-r    #will hold hundreds place
                    x=r-n    #if a number such as 21 appears, will hold tens place
                    if r in NumD10_19s:   #situations such as 'one hundred twelve'
                        first=NumD100s[s] 
                        second=NumD10_19s[r]
                        return first+'_'+second 
                    
                    if r in NumD10s and n not in NumD1s:  #situtations such as 'one hundred twenty'
                        first=NumD100s[s]
                        second=NumD10s[r]
                        return first+'_'+second

                    if r in NumD1s and n in NumD1s:  #situtations such as 'one hundred one'
                        first=NumD100s[s]
                        second=NumD1s[n]
                        return first+'_'+second
                    
                    if x in NumD10s and n in NumD1s: #situations such as 'one hundred twenty one'
                        
                        first=NumD100s[s]
                        second=NumD10s[x]
                        third=NumD1s[n]
                        return first+'_'+second+'_'+third                    
                    
                        

    def __add__(self,other):
        '''add two integers'''
        return TinyNumber((self.value + other.value)%1000)

    
    
    def __eq__(self,other):
        '''check if two integers are equivalent'''
        return self.value==other.value
```
